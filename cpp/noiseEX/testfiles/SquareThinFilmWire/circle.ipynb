```python
def writeCirclePts(r,N,z,charLen):
    print("\n\n\n")
    cl_1 = []
    ptList=[]
    theta = np.linspace(0,2*np.pi,N)
    co = np.array([r*np.cos(theta),r*np.sin(theta),np.zeros(N)]).T
    for i,c in enumerate(co):
        ptList.append(gmsh.model.geo.add_point(c[0],c[1],z,charLen))

    for i,tag in enumerate(ptList):
        if(i >= 1):
            print(ptList[i-1],ptList[i])
            cl_1.append(gmsh.model.geo.add_line(ptList[i-1],ptList[i]))
    gmsh.model.geo.removeAllDuplicates()
    print("\n\n\n")
    return cl_1 , ptList
```

```python
gmsh.initialize()

gmsh.model.add("t")



charLen = 0.09
N = 20
R = 2
W = 0.5
z_off = 1 # thickness

cl_1, pts_1 = writeCirclePts(R,N,-z_off/2,charLen)
t1 = gmsh.model.geo.add_curve_loop(cl_1)
cl_2, pts_2 = writeCirclePts(R-W,N,-z_off/2,charLen)
t2 = gmsh.model.geo.add_curve_loop(cl_2)

cl_3, pts_3 = writeCirclePts(R,N,z_off/2,charLen)
t3 = gmsh.model.geo.add_curve_loop(cl_3)
cl_4, pts_4  = writeCirclePts(R-W,N, z_off/2,charLen)
t4 = gmsh.model.geo.add_curve_loop(cl_4)

line_vert = []
for pt1, pt2, pt3, pt4 in zip(pts_1,pts_2,pts_3,pts_4):
    line_vert.append(gmsh.model.geo.add_line(pt1,pt3))


for i in range(0,len(line_vert)-1):
    vert1 = gmsh.model.geo.add_curve_loop([line_vert[i],cl_1[i],line_vert[(i+1)%(len(line_vert) - 1)],cl_3[i]],reorient=True)
    surf3 = gmsh.model.geo.add_plane_surface([vert1])









surf1 = gmsh.model.geo.add_plane_surface([t1,t2])
surf2 = gmsh.model.geo.add_plane_surface([t3,t4])



gmsh.model.geo.synchronize()
gmsh.fltk.run()

gmsh.finalize()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res = pd.read_csv("stevecomp.csv")
```

```python
res_rm = res[res["remiter"] == 2].copy()
res_rm["W"] = res_rm["W"]*1e6
sns.relplot(data=res_rm,x="W",y="numerical")

plt.xscale("log")
plt.yscale("log")
```
